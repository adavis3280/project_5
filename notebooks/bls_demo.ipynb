```python
import requests
import json
import prettytable
import pandas as pd
import bls
from bls_datasets import oes, qcew
```

```python
FIPS_list=[]
year_list=range(2014,2019,1)
```

```python
df_qcew = qcew.get_data('industry', rtype='dataframe', year='2014', qtr='1', industry='10')
cols=df_qcew.columns
```

```python
# desired format for pulling the data
df=pd.DataFrame(columns=cols)
for i in set(year_list):
    for j in (1,2,3,4):
        df_qcew = qcew.get_data('industry', rtype='dataframe', year=str(i), qtr=str(j), industry='10')
        df=pd.concat([df,df_qcew])
df.shape        
```
